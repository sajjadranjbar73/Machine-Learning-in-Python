```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Automobile.csv')
data.head()
```

```python
sns.distplot(data['wheel_base'])
```

```python
data['wheel_base'].isnull().sum()
```

```python
data.isnull().sum()
```

```python
sns.distplot(data['price'])
```

```python
sns.distplot(data['horsepower'])
```

```python
sns.jointplot(data['horsepower'], data['price'])
```

```python
sns.jointplot(data['horsepower'], data['wheel_base'])
```

```python
data.corr()
```

```python
sns.jointplot(data['horsepower'], data['wheel_base'], kind='hex')
```

```python
sns.pairplot(data)
```

```python
sns.swarmplot(data['fuel_type'], data['horsepower'])
```

```python
sns.boxplot(data['number_of_doors'], data['horsepower'], hue=data['fuel_type'])
```

```python
sns.barplot(data['number_of_doors'], data['horsepower'], hue=data['fuel_type'])
```

```python
sns.barplot(data['number_of_doors'], data['horsepower'], hue=data['engine_location'])
```
